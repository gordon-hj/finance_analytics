```python
def read_to_expand(filename):
    data = pd.read_csv(filename)
    expanded = pd.concat([
        data, 
        data['WT3M_EXPR_LSR_CLTUR_VALUE'].str.split(pat=',', expand=True).stack().reset_index().set_index('level_0')
    ], axis=1)
    del expanded['WT3M_EXPR_LSR_CLTUR_VALUE']
    del expanded['ONE_YEAR_WITHIN_EXPR_LSR_CLTUR_VALUE']
    expanded['hobby'] = expanded[0]
    del expanded[0]
    del expanded['level_1']
    return expanded
```

```python
from datetime import datetime 
```

```python
yearmonth = pd.to_datetime("2023:10", format="%Y:%m")
```

```python
yearmonths = [(yearmonth - pd.DateOffset(months=i)).strftime("%Y%m") for i in range(24)]
```

```python
result = pd.DataFrame()
results = []

for ym in yearmonths:
    expand = read_to_expand('CI_LSR_CLTUR_LCLAS_TY_ACCTO_EXPR_AT_INFO_' + ym + '.csv')
    results.append(expand)
```

```python
result = pd.concat(results)
```
